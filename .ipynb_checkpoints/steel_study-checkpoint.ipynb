## Исследовательский анализ

#### data_arc.csv — данные об электродах

Для каждой партии рассчитаем электроэнергию, умножив мощность на время воздействия и рассмотрим распределение затраченного электричества по партиям.

```python
# создадим новый столбец с воздействием электродов за промежуток времени (мощность * секунды)
data_arc['time'] = data_arc['arc_end'] - data_arc['arc_beg']
data_arc['power_by_time'] = data_arc['time'].array.seconds * data_arc['active_power']
```

```python
#подсчитаем суммы для каждой партии и визуализируем распределение
data_arc_key = data_arc.pivot_table(index = 'key', values = 'power_by_time', aggfunc = 'sum').sort_values(by='power_by_time', ascending=False)
display(data_arc_key.describe())
display(data_arc_key.tail())
data_arc_key.hist(bins = 100, range = (0, 4000))
plt.show()
data_arc_key.boxplot()
plt.show()
```

```python
# определим величину и индекс выброса
display(data_arc_key['power_by_time'].max(), data_arc_key['power_by_time'].idxmax())
```

#### data_temp.csv — результаты измерения температуры

```python
# Отберем только подходящие партии и рассмотрим распределение финальной температуры.
# найдем партии, в которых одно измерение
data_temp_count = data_temp.pivot_table(index = 'key', values =['time'], aggfunc = 'count').query('time >= 2')
data_temp_count
```

```python
#создадим список с измерениями более 1 раза
list_temp_key_out = data_temp_count.index.tolist()
data_temp_new = data_temp.query('key in @list_temp_key_out')
data_temp_new
```

```python
# найдем последние измерения - наш целевой признак

data_temp_max_time = data_temp_new.pivot_table(index = 'key', values =['time'], aggfunc = ('max'))
data_temp_max_time
```

```python
# добавим температуру
data_final_temp = data_temp_max_time.merge(data_temp_new, on = 'time', how = 'left')
data_final_temp
```

```python
#найдем начальную температуру
data_temp_min_time = data_temp_new.pivot_table(index = 'key', values =['time'], aggfunc = ('min'))
data_temp_min_time
```

```python
# добавим температуру
data_temp_min_time = data_temp_min_time.merge(data_temp_new, on = 'time', how = 'left')
data_temp_min_time
```

```python
# соберем таблицу с начальной и финальной температурами
data_final_temp = data_final_temp.merge(data_temp_min_time, on = 'key', how = 'left')
data_final_temp
```

```python
# уберем время,переименуем столбцы
data_final_temp=data_final_temp.rename(columns={'temp_y':'temp_beg', 'temp_x':'temp_end'})
# data_final_temp = data_final_temp.drop(['time_x', 'time_y'], axis = 1)
data_final_temp
```

#### data_wire_time.csv и data_bulk_time.csv 

Найдем последнее время добавок, для сравнения с последним измерением температуры

```python
# создадим столбец с последним временем в data_wire_time
for i in range(len(data_wire_time)):
     data_wire_time.loc[i, 'wire_last_time'] = max(data_wire_time.iloc[i, 1:9])
        
data_wire_time
```

```python
# создадим столбец с последним временем в data_bulk_time
for i in range(len(data_bulk_time)):
     data_bulk_time.loc[i, 'bulk_last_time'] = max(data_bulk_time.iloc[i, 1:15])
        
data_bulk_time

```

```python
# объеденим таблицы с временем для проверки
data_final_temp_wire = data_final_temp.merge(data_wire_time.loc[:, ['key', 'wire_last_time']], on = 'key', how = 'left')
data_final_temp_wire_bulk = data_final_temp_wire.merge(data_bulk_time.loc[:, ['key', 'bulk_last_time']], on = 'key', how = 'left')
data_final_temp_wire_bulk
```

```python
# произведем проверку, что замер финальной температуры сделан после добавок
data = data_final_temp_wire_bulk[data_final_temp_wire_bulk['time_x']>=(data_final_temp_wire_bulk['wire_last_time'])]
data
```

## Построение модели

### Выборки

```python
# разделим исходные данные на обучающую и тестовую выборки
data_train, data_test = train_test_split(data, test_size=0.15, random_state=12345)
```

```python
# создадим переменные с признаками и целевыми признаками из выборок
features_train = data_train.drop(['temp_end'], axis = 1)
target_train = data_train['temp_end']

features_test = data_test.drop(['temp_end'], axis = 1)
target_test = data_test['temp_end']
```

```python
# проверим размерность
display(features_train.shape, target_train.shape, features_test.shape, target_test.shape)
```

### Обучение моделей

```python
# Линейная регрессия 
model = LinearRegression()
scores = cross_val_score(model, features_train, target_train, scoring = 'neg_mean_absolute_error', cv=5) 
final_score = scores.mean()*(-1)

print(final_score)
```

```python
# Случайный лес в регрессии (переберем глубину, количество деревьев и число блоков для кросс-валидации)

best_model = None
best_result = 7
best_est = 0
best_depth = 0
best_i = 0
for est in (100, 200):
    for depth in (5, 10):
        for i in (5, 10):
            model = RandomForestRegressor(random_state=12345, n_estimators=est, max_depth=depth) 
            scores = cross_val_score(model, features_train, target_train, scoring = 'neg_mean_absolute_error', cv=i) 
            final_score = scores.mean()*(-1)
            if final_score < best_result:
                best_model = model
                best_result = final_score
                best_est = est
                best_depth = depth
                best_i = i

print((best_result), best_est, best_depth, best_i )
```

```python
# Градиентный бустинг LightGBM
model = lgb.LGBMModel(objective="regression")

param_grid = {
    'n_estimators': [500, 1000],
    'learning_rate': [0.005, 0.01],
    'max_depth': [15,20,25]  
}

gs = GridSearchCV(
        estimator=model,
        param_grid=param_grid, 
        cv=10, 
        n_jobs=-1, 
        scoring='neg_mean_absolute_error',
        verbose=2
    )
    
fitted_model = gs.fit(features_train, target_train)
  
```

```python
print(gs.best_score_*(-1)) 
```

```python
print(gs.best_params_)
```

```python
# CatBoost
train_dataset = cb.Pool(features_train, target_train) 
test_dataset = cb.Pool(features_test, target_test)
model_cb = CatBoostRegressor(loss_function='MAE')

grid = {'iterations': [ 500, 1000],
        'learning_rate': [0.01, 0.1],
        'depth': [4, 10, 20]}
model_cb.grid_search(grid, train_dataset)
model_cb.get_best_score()
```

```python
model_cb.get_all_params()
```
